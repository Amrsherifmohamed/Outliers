```python


import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
#con = sqlite3.connect(".\\OutliersAPP.db")
df = pd.read_sql_query("SELECT * FROM PlayLists ", con)

# Verify that result of SQL query is stored in the dataframe
# print(df.head())

#con.close()
```

```python


movies_df = df

movies_df.head()
```

```python
movies_df['Name'] = movies_df['Name'].apply(lambda x: x.strip())
movies_df.head()
```

```python
movies_df['Category'] = movies_df.Category.str.split('|')
movies_df.head()
```

```python

moviesWithGenres_df = movies_df.copy()


for index, row in movies_df.iterrows():
    for genre in row['Category']:
        moviesWithGenres_df.at[index, genre] = 1

moviesWithGenres_df = moviesWithGenres_df.fillna(0)
moviesWithGenres_df.head()

```

```python
userInput = [
            {'Name':'php toutorila', 'rating':5},
            {'Name':'PHP', 'rating':3.5},
            {'Name':'c++', 'rating':2},
            {'Name':"c", 'rating':5},
            {'Name':'reactjs', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python

inputId = movies_df[movies_df['Name'].isin(inputMovies['Name'].tolist())]
#نقوم باستخدام دالة الدمج لاستخراج هوية الأفلام وإضافتها لقائمة الأفلام التي قيمها المستخدم
inputMovies = pd.merge(inputId, inputMovies)
#يمكننا التخلص من قائمة التصنيفات وسنة الإصدار لعدم حاجتنا لها حالياً
inputMovies = inputMovies.drop('Category', 1).drop('DateAdded', 1)

inputMovies
```

```python
userMovies = moviesWithGenres_df[moviesWithGenres_df['Id'].isin(inputMovies['Id'].tolist())]
userMovies
```

```python

userMovies = userMovies.reset_index(drop=True)

userGenreTable = userMovies.drop('Id', 1).drop('Name', 1).drop('Category', 1).drop('Description', 1).drop('DateAdded', 1).drop('UserId', 1)
userGenreTable

```

```python
inputMovies['rating']
```

```python
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])

userProfile
```

```python
#نستخرج التصنيفات من مصفوفة الأفلام ونخزنها في dataframe
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['Id'])
#نتخلص من كل الخصائص الغير مهمة في الوقت الحالي
genreTable = genreTable.drop('Id', 1).drop('Name', 1).drop('Category', 1).drop('Description', 1).drop('DateAdded', 1).drop('UserId', 1)
genreTable.head()
```

```python
genreTable.shape
```

```python
#نضرب قيم كل تصنيف في قائمة الأفلام بالوزن من ملف المستخدم ونستخلص المتوسط 
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```python
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)

recommendationTable_df.head()
```
